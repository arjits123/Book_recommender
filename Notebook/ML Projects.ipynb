```python
year_corrections = {
    'DK Publishing Inc': 2000,
    'Gallimard': 2003
}
```

```python
Q1 = books['Year-Of-Publication'].quantile(0.25)
Q3 = books['Year-Of-Publication'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
books = books[(books['Year-Of-Publication'] >= lower_bound) & (books['Year-Of-Publication'] <= upper_bound)]
```

## Collaborative Filtering Method 

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
# users indexes who have rated more than 200 books
rating_200_above = ratings_with_books.groupby('User-ID').count()['Book-Rating'] > 200
y = rating_200_above[rating_200_above == True].index
```

```python
filtered_rating_df = ratings_with_books[ratings_with_books['User-ID'].isin(y)]
```

```python
filtered_rating_df.head()
```

```python
# gives the books which has more than 50 ratings
y = filtered_rating_df.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings_df = filtered_rating_df[filtered_rating_df['Book-Title'].isin(famous_books)]
```

```python
"""
This dataframe have books with more than 50 ratings
And Users who have rated more than 200 books
"""
final_ratings_df.head()
```

```python
pt = final_ratings_df.pivot_table(index='Book-Title', columns='User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt.shape
```

```python
pt.tail()
```

```python
"""
- Every row(book) has 794 users, so we have 794 dimensional space where every book is a vector (total 697 vectors)
- We need to calculate the distance between these vectors using cosine angle/ cosine similarity
"""

from sklearn.metrics.pairwise import cosine_similarity

#distance of each book with every other book
# cosine_similarity(pt).shape

similarity_score = cosine_similarity(pt)
similarity_score[0]
```

```python
pt.index
```

```python
# np.where with single argument helpt find indices of the elements in array
# where ever the
np.where(pt.index == "1984")[0][0]
```

```python
dis = similarity_score[np.where(pt.index == "1984")[0][0]]
index_with_score = list(enumerate(dis))
sorted_list = sorted(index_with_score, key=lambda x: x[1], reverse = True)
top_5_recommendation = sorted_list[1:6]
for i in top_5_recommendation:
    print(i[0])
```

## Recommendation Function

```python
def recommend(book_name):
    
    # get the book_index
    book_index = np.where(pt.index == book_name)[0][0]
    distances = similarity_score[book_index]
    index_with_score = list(enumerate(distances))
    sorted_list = sorted(index_with_score, key=lambda x: x[1], reverse = True)
    top_5_recommendation = sorted_list[1:6]
    
    data = []
    for i in top_5_recommendation:
        items = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(items)
        
    return data
        
```

```python
books = recommend('The Da Vinci Code')
```
